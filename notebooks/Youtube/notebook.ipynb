```python
import yt_dlp
```

```python
URLS = [
    "https://www.youtube.com/watch?v=sZCr0yn1U5k",
    "https://www.youtube.com/watch?v=VGg8Hfsqhic",
    "https://www.youtube.com/watch?v=ACduemO9d2A",
    "https://www.youtube.com/watch?v=GAt3Hx6Djkk",
    "https://www.youtube.com/watch?v=E47mg2qjy_U",
    "https://www.youtube.com/watch?v=wMGkdu7K5RI",
    "https://www.youtube.com/watch?v=jV-i8jChdgE"
]
```

```python
OPTIONS = [
    {
    'format': 'bestvideo[ext=mp4]+bestaudio[ext=m4a]/best[ext=mp4]/best',  # Priorise MP4
    'merge_output_format': 'mp4',  # Fusionne en MP4 si nécessaire
    'outtmpl': '%(title)s.%(ext)s',
    },
    {
    # Prefer best video+audio, but don't force MP4/M4A
    "format": "bv*+ba/b",
    # If possible, remux the final file to mp4 (needs ffmpeg)
    "merge_output_format": "mp4",
    "postprocessors": [{"key": "FFmpegVideoRemuxer", "preferedformat": "mp4"}],
    "outtmpl": "%(title)s.%(ext)s",
    },
    {
    'format': 'bestvideo+bestaudio/best',
    'outtmpl': '%(title)s.%(ext)s',
    }
]
```

```python

for url in URLS:
    success = False  # Pour savoir si le téléchargement a réussi

    for i, ydl_opts in enumerate(OPTIONS):
        try:
            with yt_dlp.YoutubeDL(ydl_opts) as ydl:
                ydl.download([url])
            success = True
            break  # On arrête si ça a marché
        except Exception as e:
            print(f"❌ Échec avec l'option {i+1} pour {url} ({type(e).__name__} : {e})")

    if not success:
        print(f"⚠️ Impossible de télécharger : {url} après {len(OPTIONS)} essais.")
```
