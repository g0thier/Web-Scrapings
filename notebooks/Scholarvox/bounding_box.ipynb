```python
import os
import fitz  # pymupdf

def test_croped(pdf_path:str):
    print(f"\n=== Analyse de {pdf_path} ===")

    doc = fitz.open(pdf_path)
    page = doc[0] # première page

    # Taille de la page
    page_width, page_height = page.rect.width, page.rect.height

    print(f"Page size: {page_width} x {page_height}")

    # Dessins vectoriels (lignes, rectangles, etc.)
    for i, drawing in enumerate(page.get_drawings()):
        rect = drawing["rect"]        # bounding box de la forme
        x0, y0, x1, y1 = rect
        print(f"Dessin {i} : ({x0}, {y0}, {x1}, {y1}) ", f"→ w={x1 - x0}, h={y1 - y0}")
```

```python
# liste tout les fichiers PDF du dossier courant
pdf_files = [f for f in os.listdir() if f.lower().endswith('.pdf')]

for filename in pdf_files:
    test_croped(filename)
```
